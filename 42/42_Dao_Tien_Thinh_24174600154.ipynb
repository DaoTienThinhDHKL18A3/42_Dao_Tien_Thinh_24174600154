```python
class NhanVien:
    def __init__(self, ma_nv, luong_cb):
        self.__ma_nv = ma_nv
        self.__luong_cb = luong_cb

    def tinh_luong(self):
        return self.__luong_cb

    def get_ma_nv(self):
        return self.__ma_nv

    def get_luong_cb(self):
        return self.__luong_cb

class NhanVienTangCa(NhanVien):
    def tinh_luong(self):
        return self.get_luong_cb() * 1.2
    
class NhanVienChuyenChuyen(NhanVien):
    def tinh_luong(self):
        return self.get_luong_cb() + 500000

if __name__ == "__main__":
    danh_sach = [
        NhanVien("NV01", 10000000),
        NhanVienTangCa("NV02", 10000000),
        NhanVienChuyenChuyen("NV03", 10000000)
    ]

    for nv in danh_sach:
        print(f"Mã NV: {nv.get_ma_nv()} - Lương: {nv.tinh_luong():,.0f} VNĐ")
```

```python
import numpy as np

san_luong = [12, None, 8, 15, None, 10]

arr = np.array(san_luong, dtype=float)
arr = arr.reshape(2, 3)

print("Sau khi reshape:")
print(arr)

mean_value = np.nanmean(arr)     
arr[np.isnan(arr)] = mean_value

print("Sau khi thay NaN:")
print(arr)

phan_tu_lon_hon_9 = arr[arr > 9]
gia_tri_nho_nhat = np.min(phan_tu_lon_hon_9)

print("Các phần tử > 9:", phan_tu_lon_hon_9)
print("Giá trị nhỏ nhất trong tập này:", gia_tri_nho_nhat)

```

```python
import pandas as pd
df = pd.read_csv("nhan_su.csv")

df_loc = df[
    df["Ma_NV"].astype(str).str.contains("NS", case=False, na=False) &
    (df["Nam_Cong_Tac"] >= 2021)
]

print(df_loc)
```

```python
mean_thanh_tich = df_loc["Thanh_Tich"].mean()

df_loc["Thanh_Tich"] = df_loc["Thanh_Tich"].fillna(mean_thanh_tich)

print(df_loc)
```

```python
thanh_tich_theo_phong = df_loc.groupby("Phong_Ban")["Thanh_Tich"].sum()
print(thanh_tich_theo_phong)
```

```python
pivot_table = pd.pivot_table(
    df_loc,
    index="Phong_Ban",
    columns="Nam_Cong_Tac",
    values="Thanh_Tich",
    aggfunc="sum",
    fill_value=0
)

print(pivot_table)
```
